## Exercise 1: Variable Assignment and Basic Operations

Create variables to store the following geospatial data:

- The latitude and longitude of New York City: 40.7128, -74.0060.
- The population of New York City: 8,336,817.
- The area of New York City in square kilometers: 783.8.

Perform the following tasks:

1. Calculate and print the population density of New York City (population per square kilometer).
2. Print the coordinates in the format "Latitude: [latitude], Longitude: [longitude]".

```python
# **********************************************************************************************
# * Name:   exercise_01.py
# * Author: Teo Espero
# *         MAS-GIT
# * Date written:   01/17/2026
# * Description:    This is to demonstrate the use of variables and data types. For this
# *                 exercise we will use the data about New York, their lat long, area
# *                 and population size
# **********************************************************************************************


# -------------------------------------------------------
# - Declare our data types
# -------------------------------------------------------
new_york_city = {
    "name": "New York City",
    "coordinates": (-74.0060, 40.7128),  # (lon, lat)
    "population": 8_336_817,
    "area_sq_km": 783.8
}

# -------------------------------------------------------
# - Access the data and do the computations
# -------------------------------------------------------
lat, lon = new_york_city["coordinates"]
pop_density = new_york_city["population"]/new_york_city["area_sq_km"]

print("------------------------------------------------")
print("City statistics:")
print("------------------------------------------------")
print(f"City: {new_york_city['name']}")
print(f"Total Population: {new_york_city['population']:,}")
print(f"Population density: {pop_density:,.2f}")
print(f"Latitude: {lat:.5f}")
print(f"Longitude: {lon:.5f}")
print("------------------------------------------------")

# -------------------------------------------------------
# - EOF
# -------------------------------------------------------
```
